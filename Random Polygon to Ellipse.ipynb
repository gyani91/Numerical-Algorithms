```python
def plot_sequence(n,k):
    #considering n regularly spaced value between the interval 0 to 2 pi
    theta=np.linspace(0,2*np.pi,num=n+1)
    X = 10*np.cos(theta)
    Y = 10*np.sin(theta)
    
    #Calculating X_0 and Y_0
    X = X - np.mean(X)
    Y = Y - np.mean(Y)
    X = X/np.linalg.norm(X)
    Y = Y/np.linalg.norm(Y)
    #Plotting Polygon_0
    plt.plot(X,Y)
    
    #Calculating X_k and Y_k
    for j in range(k):
        for i in range(n):
            X[i] = (X[i]+X[i+1])/2
            Y[i] = (Y[i]+Y[i+1])/2
        X[n]=X[0]
        Y[n]=Y[0]
        #Plotting Polygon_k
        plt.plot(X,Y)
    plt.xlabel('Sequence with n='+ repr(n) + ' and k='+ repr(k))
    plt.show()
```
